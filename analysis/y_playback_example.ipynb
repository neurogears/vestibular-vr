```python
import numpy as np
import aeon.io.api as api
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create a range of regularly spaced times
n = 10000  # Number of times
step = 3300  # Time between each in microseconds
y_value_increment = 1  # Incremental value for the y-axis

# Generate the x-values (which are all 0)
x_values = np.zeros(n)

# Generate the y-values (which increase over time)
y_values = np.arange(0, n * y_value_increment, y_value_increment)

times = [np.datetime64(api.aeon(0)) + np.timedelta64(i, 'us') for i in np.arange(0, n*step, step)]
print(times)

# construct into a data frame
dataframe = pd.DataFrame(y_values, index=times)

dataframe.to_csv('y_playback.csv', header=False, sep=',', lineterminator='\n')


```
